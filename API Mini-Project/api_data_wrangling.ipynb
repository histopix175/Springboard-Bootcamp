```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

# Cameron Hicks Mini API - Project

I am starting this section assuming that all cells prior to this have been run, and therefore all packages imported and variables definited from above are valid.

```python
# Fetching data for both London and Seattle to compare rainfall in both cities

# Common Parameters for both cities

year_start_date = '2022-01-01'
year_end_date = '2022-12-31'

# Coordinates for London
London_latitude = 51.5074
London_longitude = -0.1278

params_London = {
    'latitude': London_latitude,
    'longitude': London_longitude,
    'start_date': year_start_date,
    'end_date': year_end_date,
    'hourly': 'precipitation',
    'timezone': 'Europe/London'
}

London_response = requests.get(base_url, params=params_London)

if London_response.status_code == 200:
    print('London Data fetched successfully!')
else:
    print(f'Failed to fetch London data. Status code: {London_response.status_code}')

# Coordinates for Seattle
Seattle_latitude = 47.6062
Seattle_longitude = -122.3321

params_Seattle = {
    'latitude': Seattle_latitude,
    'longitude': Seattle_longitude,
    'start_date': year_start_date,
    'end_date': year_end_date,
    'hourly': 'precipitation',
    'timezone': 'America/Los_Angeles'
}

Seattle_response = requests.get(base_url, params=params_Seattle)

if Seattle_response.status_code == 200:
    print('Seattle Data fetched successfully!')
else:
    print(f'Failed to fetch London data. Status code: {Seattle_response.status_code}')
```

```python
# London Data Cleaning and converting to Data Frame

data_Lon = London_response.json()
hourly_data_Lon = data_Lon['hourly']

London_df = pd.DataFrame(hourly_data_Lon)
London_df['time'] = pd.to_datetime(London_df['time'])
London_df.set_index('time', inplace=True)
London_df.head()
```

```python
London_df.isnull().sum()
```

```python
London_df.isna().sum()
```

```python
# Seattle Data Cleaning and converting to Data Frame

data_Sea = Seattle_response.json()
hourly_data_Sea = data_Sea['hourly']

Seattle_df = pd.DataFrame(hourly_data_Sea)
Seattle_df['time'] = pd.to_datetime(Seattle_df['time'])
Seattle_df.set_index('time', inplace=True)
Seattle_df.head()
```

```python
Seattle_df.isnull().sum()
```

```python
Seattle_df.isna().sum()
```

```python
# Combining the two DataFrames

London_df = London_df.rename(columns={"precipitation": "London Precipitation"})
Seattle_df = Seattle_df.rename(columns={"precipitation": "Seattle Precipitation"})

combined_df = pd.concat([London_df, Seattle_df], axis=1)

combined_df.head()
```

```python
# Converting the hour data to weeks

combined_weekly_df = combined_df.resample('W-SAT').sum()
combined_weekly_df.index.name = 'week end date'
combined_weekly_df.head(20)
```

```python
# Data Cleaning - dropping the first row since it is not a full week of data, only one day, and very low precipitation

combined_weekly_df = combined_weekly_df.drop(combined_weekly_df.index[0])
```

```python
combined_weekly_df.head()
```

```python
# Ploting the data on a bar graph

combined_weekly_df.plot(kind='bar', figsize=(20, 12))
plt.title('Total 2022 Weekly Precipitation: London vs New York')
plt.xlabel('Week End Date')
plt.ylabel('Precipitation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Showing the weekly accumulative precipitation for each city to determine which city had the most precipitation without a doubt

cumlative_df = combined_weekly_df[['London Precipitation', 'Seattle Precipitation']].cumsum()
cumlative_df.plot(kind='bar', figsize=(20, 12))
plt.title('Accumulative 2022 Precipitation per week: London vs New York')
plt.xlabel('Week End Date')
plt.ylabel('Precipitation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Conclusion

Seattle experienced significatly more precipitation than London in 2022, especially during weeks in winter and spring. It is clear from the Total 2022 Weekly Precipitation chart that Seattle experienced many weeks of significantly more precipitation, with London only experiencing higher volumns in approximately 18 weeks. We can see form the Accumulative 2022 Precipitation per week chart that London's precipitation was nearly half of Seattle's for the year.
