```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
travel_data = pd.read_csv("Travel details dataset.csv")
```

```python
travel_data.head()
```

```python
travel_data.shape
```

```python
travel_data.info()
```

```python
dup = travel_data.duplicated().sum()

print(f"There are {dup} duplicated values")
```

```python
travel_data.isna().sum()
```

```python
# There are 3 rows of the 139 records that have NA values. Droping those rows to have correct information.

travel_data_cleaned = travel_data.dropna()
```

```python
travel_data_cleaned.shape
```

```python
travel_data_cleaned.dtypes
```

```python
# I want to compare the travel data for each year. To do that I will add a year column for the start date of the trip. 

travel_data_cleaned['Start date'] = pd.to_datetime(travel_data_cleaned['Start date'])
travel_data_cleaned['End date'] = pd.to_datetime(travel_data_cleaned['End date'])

travel_data_cleaned['Start year'] = travel_data_cleaned['Start date'].dt.year
travel_data_cleaned['End year'] = travel_data_cleaned['End date'].dt.year

travel_data_cleaned.head()
```

```python
print(travel_data_cleaned[travel_data_cleaned['Start year'] != travel_data_cleaned['End year']].index)
```

```python
travel_data_cleaned.dtypes
```

```python
travel_data_cleaned['Transportation type'].unique()
```

```python
travel_data_cleaned['Transportation type'] = travel_data_cleaned['Transportation type'].replace('Flight', 'Airplane')
travel_data_cleaned['Transportation type'] = travel_data_cleaned['Transportation type'].replace('Plane', 'Airplane')
travel_data_cleaned['Transportation type'] = travel_data_cleaned['Transportation type'].replace('Car rental', 'Car')
```

```python
travel_data_cleaned['Transportation type'].unique()
```

```python
df = travel_data_cleaned

df.head()
```

```python
df['Start year'] = df['Start year'].astype(str)
```

```python
df['Start year'].unique()
```

```python
df['Duration (days)'].unique()
```

```python
df['Traveler age'].unique()
```

```python
df['Traveler nationality'].unique()
```

```python
df['Transportation type'].unique()
```

```python
df['Traveler age'].hist(bins=30)
```

```python
# I'm going to explore the travel trends for age groups.

bins = [0, 25, 35, 50, 100]
labels = ['Under 25', '25–34', '35–49', '50+']

df['Age Group'] = pd.cut(df['Traveler age'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
df['Age Group'].value_counts().sort_index().plot(
    kind='bar',
    title='Number of Trips by Age Group'
)
```

```python
df.groupby('Start year')['Trip ID'].count().plot(
    kind='bar',
    title='Number of Trips by Year'
)
```

```python
df.groupby(['Start year', 'Age Group'])['Trip ID'] \
  .count() \
  .unstack() \
  .plot(kind='line', marker='o', title='Trips by Age Group Over Time')
```

```python
sns.boxplot(
    x='Age Group',
    y='Duration (days)',
    data=df
)
```

```python
top_nationalities = df['Traveler nationality'].value_counts().head(5).index

sns.boxplot(
    x='Traveler nationality',
    y='Duration (days)',
    data=df[df['Traveler nationality'].isin(top_nationalities)]
)
```

```python
transport_age = (
    df.groupby(['Age Group', 'Transportation type'])
      ['Trip ID']
      .count()
      .unstack()
      .fillna(0)
)

transport_age.plot(
    kind='bar',
    stacked=True,
    title='Transportation Type by Age Group'
)
```

```python
transport_modes = df['Transportation type'].unique()
age_order = labels

for mode in transport_modes:
    subset = df[df['Transportation type'] == mode]

    subset.groupby(['Start year', 'Age Group'])['Trip ID'] \
      .count() \
      .unstack() \
      .plot(kind='line', marker='o', title=f'{mode} - Trips by Age Group Over Time')
    plt.show()
```

```python
transport_modes = df['Transportation type'].unique()
age_order = labels

for mode in transport_modes:
    subset = df[df['Transportation type'] == mode]

    subset.groupby(['Start year', 'Traveler nationality'])['Trip ID'] \
      .count() \
      .unstack() \
      .plot(kind='line', marker='o', title=f'{mode} - Trips by Age Group Over Time')
    plt.show()
```
