```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Aggregate the Data

```python
login_counts = df.resample('15min').size()

df_login_counts = login_counts.to_frame(name='login_count').reset_index()
```

```python
df_login_counts.head()
```

```python
df_login_counts.set_index('login_time', inplace=True)
```

## Explore the Data

```python
plt.figure(figsize=(15, 6))
plt.plot(df_login_counts.index, df_login_counts['login_count'], alpha=0.3, label='Raw 15-min')
plt.plot(df_login_counts.index, df_login_counts['login_count'].rolling(window=96).mean(), color='red', label='Daily Rolling Avg')
plt.title('Login Demand Over Time')
plt.legend()
plt.show()
```

In the total data visualization we can see the is a clear pattern in both the raw data and the daily rolling average with approximately 4 peaks per month that are immeaditely followed by a valley. The peaks and valleys appear to be growing overtime. There also appears to be an outlier around 1970-03-01 with a major spike in the Raw 15-min data, but a very similar patter in the average data. 

```python
df_login_counts.sort_values(by='login_count', ascending=False).head()
```

```python
# View average login per Day of Week
df_login_counts['weekday'] = df_login_counts.index.dayofweek
weekly_pattern = df_login_counts.groupby('weekday')['login_count'].mean()
weekly_pattern.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
weekly_pattern.plot(kind='bar', title='Average Logins by Day of Week')
```

Viewing the average login's per day of week we can see that the pattern of peaks and valleys viewed in the total dataset is dur to higher volumn of logins on weekends vs week days. Saturday see's the highest volumn with Sunday as the second highest, then Monday drops to the lowest volumn

```python
# View average login per hour of the day
daily_cycle = df_login_counts.groupby(df_login_counts.index.hour).mean()

daily_cycle.plot(kind='bar', title='Average Logins by Hour of Day')
plt.xlabel('Hour (24h format)')
plt.ylabel('Avg Logins')
```

Viewing the average number of logins by hour of day shows us that the evening and early morning hours experience the highest volumn of logins with another spike in mid day hours. This could be effected by the day of week, maybe work hours are why the day time hours experience less logins with the exception of common lunch hours.

```python
df_login_counts['hour'] = df_login_counts.index.hour
df_login_counts['day_of_week_name'] = df_login_counts.index.day_name()
```

```python
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
for day in days_order:
    day_data = df_login_counts[df_login_counts['day_of_week_name'] == day]
    hourly_avg = day_data.groupby('hour')['login_count'].mean()

    plt.figure(figsize=(8,4))
    plt.title(f"Avg Logins by Hour - {day}")
    hourly_avg.plot(kind='bar')
```

Here we can see a clear difference in pattern between weekdays which follow the same double bottom pattern, vs weekends which are closer to a roller coaster pattern. 

```python
df_login_counts.head()
```

```python
heatmap_data = (
    df_login_counts
    .groupby(['day_of_week_name', 'hour'])['login_count']
    .mean()
    .reset_index()
    .pivot(index='day_of_week_name', columns='hour', values='login_count')
)


heatmap_data = heatmap_data.reindex(days_order)

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='Reds')
plt.title('Average Logins by Day of Week and Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

This heatmap shows a clear visualization of the times of day that experience the most logings. We can see clear hot zones mid day and evenings on weekdays, and the most logins on weekends in the early hours between midnight at 5am. 

```python
df_login_counts['is_weekend'] = df_login_counts['weekday'].isin([5, 6])
```

```python
hourly_weekday = (
    df_login_counts[~df_login_counts['is_weekend']]
    .groupby('hour')['login_count']
    .mean()
)

hourly_weekend = (
    df_login_counts[df_login_counts['is_weekend']]
    .groupby('hour')['login_count']
    .mean()
)

plt.figure(figsize=(10, 5))
plt.plot(hourly_weekday.index, hourly_weekday.values, label='Weekday')
plt.plot(hourly_weekend.index, hourly_weekend.values, label='Weekend')
plt.title('Average Hourly Logins: Weekday vs Weekend')
plt.xlabel('Hour of Day')
plt.ylabel('Average Logins')
plt.xticks(range(24))
plt.legend()
plt.show()
```

This line plot visualization shows us the average number of logins per hour by day types - Weekday vs Weekend. We can see that the two types of day have very different shapes that nearly have opposite trends with the exception of a spike at the 22 hour mark. We can determine that day type has an obvious correlation with number of logins.

```python
hourly_quantiles = (
    df_login_counts
    .groupby('hour')['login_count']
    .quantile([0.25, 0.5, 0.75])
    .unstack()
)

plt.figure(figsize=(10, 5))
plt.plot(hourly_quantiles.index, hourly_quantiles[0.5], label='Median')
plt.fill_between(
    hourly_quantiles.index,
    hourly_quantiles[0.25],
    hourly_quantiles[0.75],
    alpha=0.3,
    label='IQR'
)
plt.title('Hourly Login Distribution')
plt.xlabel('Hour of Day')
plt.ylabel('Login Count')
plt.xticks(range(24))
plt.legend()
plt.show()
```

When reviewing the Interquartile Range vs the mean, we can see that it is heavily influenced by the weekday pattern, and the times with the most logins also have the least predictability. By separating the weekday and weekends in the IQR visualization below we can see that there is far more predictability between the day types, with weekday generally being more predictable. 

```python
# code for quartiles

WD_hourly_quantiles = (
    df_login_counts[~df_login_counts['is_weekend']]
    .groupby('hour')['login_count']
    .quantile([0.25, 0.5, 0.75])
    .unstack()
)

WE_hourly_quantiles = (
    df_login_counts[df_login_counts['is_weekend']]
    .groupby('hour')['login_count']
    .quantile([0.25, 0.5, 0.75])
    .unstack()
)

plt.figure(figsize=(10, 5))
plt.plot(WD_hourly_quantiles.index, WD_hourly_quantiles[0.5], color='blue', label='Weekday Median')
plt.fill_between(
    WD_hourly_quantiles.index,
    WD_hourly_quantiles[0.25],
    WD_hourly_quantiles[0.75],
    alpha=0.3,
    color='lightblue',
    label='Weekday IQR'
)
plt.plot(WE_hourly_quantiles.index, WE_hourly_quantiles[0.5], color='orange', label='Weekend Median')
plt.fill_between(
    WE_hourly_quantiles.index,
    WE_hourly_quantiles[0.25],
    WE_hourly_quantiles[0.75],
    alpha=0.3,
    color='yellow',
    label='Weekend IQR'
)
plt.title('Hourly Login Distribution')
plt.xlabel('Hour of Day')
plt.ylabel('Login Count')
plt.xticks(range(24))
plt.legend()
plt.show()
```

```python
# Growth
df_login_counts['week_number'] = df_login_counts.index.isocalendar().week

weekly_counts = df_login_counts.groupby('week_number')['login_count'].sum()
weekly_counts.plot(kind='bar', title='Total Logins by Week')
plt.xlabel('Week Number')
plt.ylabel('Total Logins')
plt.show()
```

There is a clear growth in the number of logins from the first 5 weeks to the later 5 weeks. Week 16 appears to have a very small number of logins, possibly due to only a partial week of logins tracked.

# Part 2 ‐ Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.
1) What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
2) Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:<br>
a) how you will implement the experiment<br>
b) what statistical test(s) you will conduct to verify the significance of the
observation<br>
c) how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

## Answer

1. The key metric to determine success of this experiment would be to measure <b>Drivers Active in Both Cities</b>. This metric would be measured by tracking the number of drivers who completed at least one pickup/dropoff in both Gotham and Metropolis during a shift. Success in this experiment would be to achieve a significant increase of drivers who are active in both cities during one shift proving that the toll reimbursement policy is effective in encouraging drivers to cross the city boundries. However, there are two additional metrics that would be vital to track as part of this experiment to determine if this business plan meets the Ultimate managers needs. <b>Profit</b> - The experiment would need to track the total cost of the reimbursed tolls as well as the revenue generated from all of the rides. These two will determin the profitability of this business model. <b>Rider wait times</b> - tracked to determine if the new business model will increase or decrease rider wait times. These two additional metrics are vital to incorporate in the final report of this experienment because while the experiment may be a success in encouraging drivers to be active in both cities, if it reduces profitability or increases rider wait time it may ultimately have a negative effect and not be the result Ultimate managers are looking for.

2. <b>Hypothesis</b> - Can we increase the number of drivers who are active in both Gotham and Metropolis during a shift by reimbursing drivers toll costs?<br>
To test this hypothesis, I would conduct a randomized controlled trial. Firstly, I would separate all of the Ultimate drivers into groups. Depending on existing levels of data availibility, I would group the drivers by their activity and driving levels (depending on the data this could be average number of hours worked per week, average number of rides completed per week, or possibly city preferred to drive in). Each group would then be randomly split in half with 50% receiving the toll reimbursement policy and the other 50% not receiving or being informed of the policy. If there is no data to separate into subgroups, the full driver roster would be randomly split into these two halves. Once split, data would be tracked for a minimum of 1 month. Specific metrics to be tracked (if possible):

    - Drivers Location upon beginning of shift - records city shift began in 
    - Drivers location upon end of shift - records city shift ended in
    - Driver Toll Reimbursable - boolean indicating if the driver is eligible for Toll Reimbursement or not
    - Every pick up location during the shift - records city of each pick up
    - Every drop off location during the shift - records city of each drop off
    - Cross City Line to Pick Up - boolean indicating True if the driver had to cross city line from previous drop off to next pick up
    - Cross City Line to Drop Off - boolean indicating True if the driver had to cross city line from pick up to drop off
    - Trip Revenue - records the total revenue from each trip
    - Trip Toll - boolean indicating if there was a toll during the trip
    - Pick Up Toll - boolean indicating if there was a toll between previous drop off location and next pick up location
    - Total Toll Cost - records the total cost of the toll
    - Profit - records the total revenue minus the total toll cost
    - Rider Wait Time - records how long the rider waited between requsting the ride and pick up
    - Ride Length (Time) - records the duration of the ride
    - Ride distance (Miles) - records the number of miles of the trip
    - Pick Up Distance (Miles) - records the number of miles between the previous drop off location and next pick up location
    - <b>Active in Both Cities</b> - boolean, True if any of these conditions are True: Drivers Location upon beginning of shift does not equal Drivers location upon end fo shift, Cross City Line to Pick Up is True, Cross City Line to Drop Off is True, Trip Toll is True, Pick Up Toll is True, Total Toll Cost does not equal 0.
<br>
<b>Statiscial tests</b> - Once the experiment concludes or has been ongoing for at least a month I would conduct a <b>two-proportion z-test</b> to compare the proportion of drivers active in both cities between the treatment group and the control group. If we observe significantly more drivers in the treatment group that are active in both cities we can determine the experiment was a success in the target metric. However, to test the caveats, I would then perform <b>two-sample t-tests</b> on profit and average rider wait times to determine if there was a noticable impact on either of those two metrics. Based on results from those tests, I would likely present one of four possible outcomes:<br>
- <b>Outcome 1:</b> Significant increase in cross-city drivers with neutral or positive profit and rider experience impact. Recommendation: Roll out the policy company wide to further increase driver activity.<br>
- <b>Outcome 2:</b> Significant increase in cross-city drivers with negative impact to profit. Recommendation: While the experiment was successful in increasing drivers active in both cities, the toll reimbursements are negatively impacting profits. Consider the implications of the draw backs before deciding to continue or expand the policy. If dual-city drivers are highly needed, consider continuing the program for highly active/high earning drivers only.<br>
- <b>Outcome 3:</b> Significant increase in cross-city drivers with negative impact to rider wait times. Recommendation: While the experiment was successful in increasing drivers active in both cities, the rider wait times were increased by {x} amount. Rolling out the toll reimbursement policy to all drivers could have further negative effects on rider wait time potentailly leaving the city with less demand with few to no drivers near by. Consider reducing the number of drivers eligible for the toll reimbursement policy to maintain a par of drivers available in each city, while keeping a par of drivers who are able to commute between cities as needed.<br>
<b>Outcome 4:</b>Decrease or no impact to cross-city drivers. Recommendation: The experiment showed no improved results of driver behavior becoming active in both cities. Roll back the toll reimbursement program and consider other encentives that could encourage drivers to cross city lines.
<br>
<br>
<b>Caveats & Limitations</b> - There are some caveats to consider prior to full adoption of the recommendation:<br>
- <b>Experiment Window</b> - There could be seasonal effect on the outcomes that were not made apparent in this sort term experiment. Consider an extended roll out period to reduce surprises over long term.<br>
- <b>Major Events</b> - If there is a major event likely to produce high demand in one of the cities, drivers could be disproportionatly distributed to one of the two cities, leaving customers without a nearby driver and a long wait time

## Create Target Variable - Retained

The target variable for this project is if the user was retained, measured by if the user has taken a ride in the 30 days prior to the most recent data pull. Since the data pull date was not provided, we will base it on the max trip date recorded in the data

```python
data_pull_date = udf['last_trip_date'].max()

udf['retained'] = (
    (data_pull_date - udf['last_trip_date']).dt.days <= 30
).astype(int)
```

```python
retention_rate = udf['retained'].mean()
retention_rate
```

Approximately 37.61% of users in the dataset were retained

## Feature Prep and Preprocessing

```python
# Encode categorical variables

udf = pd.get_dummies(
    udf,
    columns=['city', 'phone'],
    drop_first=True
)
```

```python
# Conver boolean column
udf['ultimate_black_user'] = udf['ultimate_black_user'].astype(int)
```

```python
# Remove dates to prevent leakage
udf_model = udf.drop(columns=['signup_date', 'last_trip_date'])
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X = udf_model.drop('retained', axis=1)
y = udf_model['retained']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

log_reg = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])

log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)[:, 1]

accuracy_score(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_prob)
```

```python
print(classification_report(y_test, y_pred))
```

```python
feature_importance = pd.Series(
    log_reg.named_steps['model'].coef_[0],
    index=X.columns
).sort_values()

feature_importance.sort_values(ascending=False)
```

```python
feature_importance.plot(kind='barh', figsize=(8, 10))
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]

# Metrics
accuracy_score(y_test, y_pred_rf)
```

```python
roc_auc_score(y_test, y_prob_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
rf_importance = pd.Series(
    rf.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

rf_importance.head(10)
```

```python
rf_importance.head(15).plot(
    kind='barh',
    figsize=(8, 6),
    title='Top Feature Importances (Random Forest)'
)
```
